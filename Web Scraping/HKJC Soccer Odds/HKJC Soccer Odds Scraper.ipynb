## HKJC Soccer Odds Scraper

```python
import json
import pandas as pd
import re
import requests
```

```python
# HAD url
HAD_url = 'https://bet.hkjc.com/football/getJSON.aspx?jsontype=odds_had.aspx'
HAD_response = requests.get(HAD_url)
HAD = HAD_response.json()

# Handicap HAD url 
Handicap_url = 'https://bet.hkjc.com/football/getJSON.aspx?jsontype=odds_hha.aspx'
Handicap_response = requests.get(Handicap_url)
Handicap = Handicap_response.json()
```

```python
# Transform the json files to dataframes
HAD_df = pd.DataFrame(HAD[1]['matches'])
Handicap_df = pd.DataFrame(Handicap[1]['matches'])
```

```python
def clean_HAD_data(HAD_df):
    H_hadodds = []
    D_hadodds = []
    A_hadodds = []
    Away_Team_ID = []
    Away_Team_Name = []
    Home_Team_ID  = []
    Home_Team_Name = []

    for item in HAD_df['hadodds']:
        H = item['H']
        D = item['D']
        A = item['A']
        H = re.findall(r'@.+', H)
        D = re.findall(r'@.+', D)
        A = re.findall(r'@.+', A)
        H = re.findall(r'\d+?\.\d+',str(H))
        D = re.findall(r'\d+?\.\d+',str(D))
        A = re.findall(r'\d+?\.\d+',str(A))
        H_hadodds.append(H[0])
        D_hadodds.append(D[0])
        A_hadodds.append(A[0])

    for item in HAD_df['awayTeam']:
        ix = item['teamNameEN']
        xi = item['teamID']
        Away_Team_Name.append(ix)
        Away_Team_ID.append(xi)
    
    for item in HAD_df['homeTeam']:
        ix = item['teamNameEN']
        xi = item['teamID']
        Home_Team_Name.append(ix)
        Home_Team_ID.append(xi)
        
    dataframe = pd.DataFrame({'Match ID':HAD_df.matchID,'Last update':HAD_df.statuslastupdated,
                              'Match Day':HAD_df['matchDay'],'Match Num':HAD_df['matchNum'],
                              'Home_Team_ID':Home_Team_ID, 'Home_Team_Name': Home_Team_Name,
                              'Away_Team_ID':Away_Team_ID,'Away_Team_Name':Away_Team_Name,
                              'H_hadodds':H_hadodds, 'D_hadodds':D_hadodds, 'A_hadodds':A_hadodds })
    return dataframe
```

```python
dataframe = clean_HAD_data(HAD_df)
```

```python
def clean_HHA_data(Handicap_df):
    H_hhaodds = []
    D_hhaodds = []
    A_hhaodds = []

    for item in Handicap_df['hhaodds']:
    
        try:
            H = item['H']
            H = re.findall(r'@.+', H)
            H = re.findall(r'\d+?\.\d+',str(H))
            H_hhaodds.append(H[0])
        
            D = item['D']
            D = re.findall(r'@.+', D)
            D = re.findall(r'\d+?\.\d+',str(D))
            D_hhaodds.append(D[0])
        
            A = item['A']
            A = re.findall(r'@.+', A)
            A = re.findall(r'\d+?\.\d+',str(A))
            A_hhaodds.append(A[0])
    
        except:
            H_hhaodds.append('NA')
            D_hhaodds.append('NA')
            A_hhaodds.append('NA')
            
    dataframe_2 = pd.DataFrame({'Match ID':Handicap_df.matchID,
                                'H_HHAodds':H_hhaodds, 'D_HHAodds':D_hhaodds, 'A_HHAodds':A_hhaodds })
        
    return dataframe_2
```

```python
dataframe_2 = clean_HHA_data(Handicap_df)
```

```python
final = pd.merge(dataframe, dataframe_2, how='left', on=['Match ID']) 
```
