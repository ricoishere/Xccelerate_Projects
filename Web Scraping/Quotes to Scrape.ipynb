```python
from bs4 import BeautifulSoup
import requests
page = 1
URL = "http://quotes.toscrape.com/page/" + str(page)
html = requests.get(URL)
soup = BeautifulSoup(html.text, "html.parser")
text = [t.get_text() for t in soup.find_all(class_='text')]
author = [a.get_text() for a in soup.find_all(class_='author')]
```

```python
page_num = 1
scrape_results = []

while True:
    page = requests.get(f'http://quotes.toscrape.com/page/{page_num}').text
    soup = BeautifulSoup(page, "html.parser")
    if not soup.body.find_all(class_='quote'):
        break
    for elem in soup.body.find_all(class_='quote'):
        this_quote = {'text': elem.span.text, 'author':elem.small.text, 'tags':[]}
        for subelem in elem.find_all('a',class_='tag'):
            this_quote['tags'].append(subelem.text)
        scrape_results.append(this_quote)
    page_num += 1

scrape_results     
```
