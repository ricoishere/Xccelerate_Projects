```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import os
import pandas as pd
import requests
import re
```

```python
url='https://www.quora.com/topic/Deep-Web'
driver = webdriver.Chrome(executable_path='/Users/yanhong/Downloads/chromedriver')
driver.get(url) 

SCROLL_PAUSE_TIME = 1.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    
    links = driver.find_elements_by_class_name('ui_qtext_more_link')
    for link in links:
        try:
            link.click()
            time.sleep(SCROLL_PAUSE_TIME)
            
        except:
            continue
            
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
divs = driver.find_elements_by_xpath("//div[@class='FeedStory AnswerFeedStory feed_item']")

data = []
for div in divs:
    try:
        topic = div.find_element_by_xpath("//a[@class='HoverMenu TopicNameLink topic_name']").text
        author = div.find_element_by_class_name('user').text
        question = div.find_element_by_class_name("ui_qtext_rendered_qtext").text
        answer = div.find_element_by_class_name("answer_body_preview").text
        DICT = {'topic': topic,'author':author,'question':question, 'answer': answer}
        data.append(DICT)
    except:
        continue
    
data = pd.DataFrame.from_dict(data)
```
