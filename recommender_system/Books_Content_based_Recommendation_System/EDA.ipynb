```python
import requests
from bs4 import BeautifulSoup

def get_description(isbn):
    data = requests.get(f'https://www.justbooks.co.uk/search/?isbn={isbn}&mode=isbn&st=sr&ac=qr')
    soup = BeautifulSoup(data.text, 'html.parser')
    summary = soup.findAll('div', {'id':'bookSummary'})
    des = [item.text for item in summary]
    print(des)
    return des
```

```python
sample['Description'] = sample['ISBN'].apply(lambda x: str(x))
```

```python
sample['Description'] = sample['Description'].apply(get_description)
```
